```python
import time
import numpy as np
```

```python
from bs4 import BeautifulSoup as bs
import requests

titles_list =[]
date_times_list =[]
for page in range(1,157):
    time.sleep(2)
    url = 'https://www.paxnet.co.kr/stock/report/report?menuCode=1111&currentPageNo='+str(page)+'&reportId=0&searchKey=title&searchValue='

    response = requests.get(url)
    html_text=response.text

    soup = bs(html_text, 'html.parser')

    titles = soup.select('div.board-type > ul > li > div:nth-child(2) > p > a')
    date_times = soup.select('div.board-type > ul > li > div:nth-child(5)')

    for i in titles:
        title = i.get_text()
        titles_list.append(title)
        #print(title)

    for i in date_times:
        date_time = i.get_text()
        new_date_time = date_time.replace('.', '-')
        if new_date_time=="작성일":
            continue
        date_times_list.append(new_date_time)
        #print(new_date_time)
df=pd.DataFrame()
df["titles"]=titles_list
df["date"]=date_times_list
df
```
