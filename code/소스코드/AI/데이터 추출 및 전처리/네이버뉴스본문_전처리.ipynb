본문 + 제목 합치기

```python
def combine_2rd_columns(col_1, col_2):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result

df_2016['내용'] = df_2016.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2017['내용'] = df_2017.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2018['내용'] = df_2018.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2019['내용'] = df_2019.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2020['내용'] = df_2020.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2021['내용'] = df_2021.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
df_2022['내용'] = df_2022.apply(lambda x: combine_2rd_columns(x['제목'], x['본문']), axis=1)
```

```python
df_2016.drop(columns=['제목','본문'],inplace=True)
df_2017.drop(columns=['제목','본문'],inplace=True)
df_2018.drop(columns=['제목','본문'],inplace=True)
df_2019.drop(columns=['제목','본문'],inplace=True)
df_2020.drop(columns=['제목','본문'],inplace=True)
df_2021.drop(columns=['제목','본문'],inplace=True)
df_2022.drop(columns=['제목','본문'],inplace=True)
```

날짜 별로 합치기

```python
def merge_date(df, year):
  base=pd.DataFrame(df[["날짜"]])
  base=base.drop_duplicates(["날짜"])
  base=base.reset_index(drop=True)

  dic={}
  for x in base["날짜"]:
    dic[x]=""

  for tit,dat in zip(df["내용"],df["날짜"]):
    dic[dat]+=" "+str(tit)
  
  result=pd.DataFrame([dic])
  result=result.transpose()
  result.columns=["내용"]
  result["날짜"]=result.index
  result.to_csv(path+year+"_Merged_Naver.csv",index=False)
```

```python
merge_date(df_2016,"2016")
merge_date(df_2017,"2017")
merge_date(df_2018,"2018")
merge_date(df_2019,"2019")
merge_date(df_2020,"2020")
merge_date(df_2021,"2021")
merge_date(df_2022,"2022")
```
